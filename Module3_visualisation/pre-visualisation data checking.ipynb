## read in visualisation data

```python
filename='visualisation_data.csv'
df=pd.read_csv(filename,
               parse_dates=['Date','SubmittedAt','AuthorisedAt','AcceptedAt','ExtractedAt',
                            'taxDueDate','motExpiryDate','dateOfLastV5CIssued','monthOfFirstRegistration'],
               nrows=None
               )
```

```python
# count of unique values and % missing data per column
pd.DataFrame({'unique value count':{col:len(df[col].unique()) for col in df},
              'missing data percentage':{col:round(df[col].isna().sum()/len(df)*100,1) for col in df},
             }
            )
```
